```python
# 1. Unzip the data to local Colab storage
!unzip -q /content/drive/MyDrive/SmartVision_AI/dataset.zip -d /content/dataset

# 2. Update data.yaml to point to Colab paths
import yaml

with open('/content/drive/MyDrive/SmartVision_AI/data.yaml', 'r') as f:
    data = yaml.safe_load(f)

# Update the paths for the Colab environment
data['path'] = '/content/dataset'
data['train'] = 'images' # Points to /content/dataset/images
data['val'] = 'images'   # Using same for val since you had one folder

with open('/content/dataset/data_colab.yaml', 'w') as f:
    yaml.dump(data, f)
```

```python
from ultralytics import YOLO

# Load the modern 'u' version of YOLOv5s
model = YOLO('yolov5su.pt')

# Train with Colab-optimized settings
results = model.train(
    data='/content/dataset/data_colab.yaml',
    epochs=50,
    imgsz=640,        # Standard resolution (much better accuracy)
    batch=32,         # Higher batch size for better stability
    device=0,         # Use the T4 GPU
    project='/content/drive/MyDrive/SmartVision_AI/results', # Saves results back to Drive
    name='smartvision_colab_run'
)
```
